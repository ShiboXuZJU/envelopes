```python
import sys
import pathlib
import os
sys.path.append(str(pathlib.Path(os.getcwd()).parent.absolute()))
from envelopes import evc, evp
import numpy as np
import matplotlib.pyplot as plt
import pickle
```

```python
wc_p = evp.WaveCache(0.5)
wc_c = evc.WaveCache(0.5)
ax_idx = 1

fig = plt.figure(figsize=(14, 10))

fig.add_subplot(4, 4, ax_idx)
evl_p = evp.Gaussian(t0=0, w=10, amp=1)
t, wave = evp.decode_envelope(evl_p, wc_p)
plt.plot(np.arange(t, t+len(wave)*wc_p.resolution, wc_p.resolution), wave, '.-', label='py')
evl_c = evc.Gaussian(t0=0, w=10, amp=1)
evl_c = pickle.loads(pickle.dumps(evl_c))
t, wave = evc.decode_envelope(evl_c, wc_c)
wave = np.array(wave, copy=False)
plt.plot(np.arange(t, t+len(wave)*wc_c.resolution(), wc_c.resolution()), wave, '.-', label='c++')
plt.legend()
plt.title(evl_p.__class__.__name__)
ax_idx += 1

fig.add_subplot(4, 4, ax_idx)
evl_p = evp.GaussianDRAG(t0=0, w=10, coef=2.5, df=0.1, amp=1, phase=0)
t, wave = evp.decode_envelope(evl_p)
plt.plot(np.arange(t, t+len(wave)*wc_p.resolution, wc_p.resolution), wave.real, '.-', label='py')
evl_c = evc.GaussianDRAG(t0=0, w=10, coef=2.5, df=0.1, amp=1, phase=0)
evl_c = pickle.loads(pickle.dumps(evl_c))
t, wave = evc.decode_envelope(evl_c, wc_c)
wave = np.array(wave, copy=False)
plt.plot(np.arange(t, t+len(wave)*wc_c.resolution(), wc_c.resolution()), wave.real, '.-', label='c++')
plt.legend()
plt.title(evl_p.__class__.__name__+' real')
ax_idx += 1

fig.add_subplot(4, 4, ax_idx)
evl_p = evp.GaussianDRAG(t0=0, w=10, coef=2.5, df=0.1, amp=1, phase=0)
t, wave = evp.decode_envelope(evl_p)
plt.plot(np.arange(t, t+len(wave)*wc_p.resolution, wc_p.resolution), wave.imag, '.-', label='py')
evl_c = evc.GaussianDRAG(t0=0, w=10, coef=2.5, df=0.1, amp=1, phase=0)
evl_c = pickle.loads(pickle.dumps(evl_c))
t, wave = evc.decode_envelope(evl_c, wc_c)
wave = np.array(wave, copy=False)
plt.plot(np.arange(t, t+len(wave)*wc_c.resolution(), wc_c.resolution()), wave.imag, '.-', label='c++')
plt.legend()
plt.title(evl_p.__class__.__name__+' imag')
ax_idx += 1

fig.add_subplot(4, 4, ax_idx)
evl_p = evp.CosineDRAG(t0=0, w=10, coef=2.5, df=0.1, amp=1, phase=0)
t, wave = evp.decode_envelope(evl_p)
plt.plot(np.arange(t, t+len(wave)*wc_p.resolution, wc_p.resolution), wave.real, '.-', label='py')
evl_c = evc.CosineDRAG(t0=0, w=10, coef=2.5, df=0.1, amp=1, phase=0)
evl_c = pickle.loads(pickle.dumps(evl_c))
t, wave = evc.decode_envelope(evl_c, wc_c)
wave = np.array(wave, copy=False)
plt.plot(np.arange(t, t+len(wave)*wc_c.resolution(), wc_c.resolution()), wave.real, '.-', label='c++')
plt.legend()
plt.title(evl_p.__class__.__name__+' real')
ax_idx += 1

fig.add_subplot(4, 4, ax_idx)
evl_p = evp.CosineDRAG(t0=0, w=10, coef=2.5, df=0.1, amp=1, phase=0)
t, wave = evp.decode_envelope(evl_p)
plt.plot(np.arange(t, t+len(wave)*wc_p.resolution, wc_p.resolution), wave.imag, '.-', label='py')
evl_c = evc.CosineDRAG(t0=0, w=10, coef=2.5, df=0.1, amp=1, phase=0)
evl_c = pickle.loads(pickle.dumps(evl_c))
t, wave = evc.decode_envelope(evl_c, wc_c)
wave = np.array(wave, copy=False)
plt.plot(np.arange(t, t+len(wave)*wc_c.resolution(), wc_c.resolution()), wave.imag, '.-', label='c++')
plt.legend()
plt.title(evl_p.__class__.__name__+' imag')
ax_idx += 1

fig.add_subplot(4, 4, ax_idx)
evl_p = evp.Triangle(t0=0, tlen=10, amp=1, fall=True)
t, wave = evp.decode_envelope(evl_p)
plt.plot(np.arange(t, t+len(wave)*wc_p.resolution, wc_p.resolution), wave, '.-', label='py')
evl_c = evc.Triangle(t0=0, tlen=10, amp=1, fall=True)
evl_c = pickle.loads(pickle.dumps(evl_c))
t, wave = evc.decode_envelope(evl_c, wc_c)
wave = np.array(wave, copy=False)
plt.plot(np.arange(t, t+len(wave)*wc_c.resolution(), wc_c.resolution()), wave, '.-', label='c++')
plt.legend()
plt.title(evl_p.__class__.__name__+' fall=True')
ax_idx += 1

fig.add_subplot(4, 4, ax_idx)
evl_p = evp.Triangle(t0=0, tlen=10, amp=1, fall=False)
t, wave = evp.decode_envelope(evl_p)
plt.plot(np.arange(t, t+len(wave)*wc_p.resolution, wc_p.resolution), wave, '.-', label='py')
evl_c = evc.Triangle(t0=0, tlen=10, amp=1, fall=False)
evl_c = pickle.loads(pickle.dumps(evl_c))
t, wave = evc.decode_envelope(evl_c, wc_c)
wave = np.array(wave, copy=False)
plt.plot(np.arange(t, t+len(wave)*wc_c.resolution(), wc_c.resolution()), wave, '.-', label='c++')
plt.legend()
plt.title(evl_p.__class__.__name__ + ' fall=False')
ax_idx += 1

fig.add_subplot(4, 4, ax_idx)
evl_p = evp.Rect(t0=0, tlen=10, amp=1)
t, wave = evp.decode_envelope(evl_p)
plt.plot(np.arange(t, t+len(wave)*wc_p.resolution, wc_p.resolution), wave, '.-', label='py')
evl_c = evc.Rect(t0=0, tlen=10, amp=1)
evl_c = pickle.loads(pickle.dumps(evl_c))
t, wave = evc.decode_envelope(evl_c, wc_c)
wave = np.array(wave, copy=False)
plt.plot(np.arange(t, t+len(wave)*wc_c.resolution(), wc_c.resolution()), wave, '.-', label='c++')
plt.legend()
plt.title(evl_p.__class__.__name__)
ax_idx += 1

fig.add_subplot(4, 4, ax_idx)
evl_p = evp.Flattop(t0=0, tlen=20, w_left=2, w_right=2, amp=1)
t, wave = evp.decode_envelope(evl_p)
plt.plot(np.arange(t, t+len(wave)*wc_p.resolution, wc_p.resolution), wave, '.-', label='py')
evl_c = evc.Flattop(t0=0, tlen=20, w_left=2, w_right=2, amp=1)
evl_c = pickle.loads(pickle.dumps(evl_c))
t, wave = evc.decode_envelope(evl_c, wc_c)
wave = np.array(wave, copy=False)
plt.plot(np.arange(t, t+len(wave)*wc_c.resolution(), wc_c.resolution()), wave, '.-', label='c++')
plt.legend()
plt.title(evl_p.__class__.__name__)
ax_idx += 1

fig.add_subplot(4, 4, ax_idx)
evl_p = evp.RippleRect(t0=0, tlen=20, w=2, amp=1, ripple0=0, ripple1=1, ripple2=0, ripple3=-0.25)
t, wave = evp.decode_envelope(evl_p)
plt.plot(np.arange(t, t+len(wave)*wc_p.resolution, wc_p.resolution), wave, '.-', label='py')
evl_c = evc.RippleRect(t0=0,
                       tlen=20,
                       w=2,
                       amp=1,
                       ripple0=0,
                       ripple1=1,
                       ripple2=0,
                       ripple3=-0.25)
evl_c = pickle.loads(pickle.dumps(evl_c))
t, wave = evc.decode_envelope(evl_c, wc_c)
wave = np.array(wave, copy=False)
plt.plot(np.arange(t, t+len(wave)*wc_c.resolution(), wc_c.resolution()), wave, '.-', label='c++')
plt.legend()
plt.title(evl_p.__class__.__name__)
ax_idx += 1

fig.add_subplot(4, 4, ax_idx)
evl_p = evp.CosineEdgeRect(t0=0, tlen=20, w=10, amp=1)
t, wave = evp.decode_envelope(evl_p)
plt.plot(np.arange(t, t+len(wave)*wc_p.resolution, wc_p.resolution), wave, '.-', label='py')
plt.legend()
plt.title(evl_p.__class__.__name__)
ax_idx += 1

fig.add_subplot(4, 4, ax_idx)
evl_p = evp.EllipseEdgeRect(t0=0, tlen=20, w=10, amp=1)
t, wave = evp.decode_envelope(evl_p)
plt.plot(np.arange(t, t+len(wave)*wc_p.resolution, wc_p.resolution), wave, '.-', label='py')
plt.legend()
plt.title(evl_p.__class__.__name__)
ax_idx += 1

fig.add_subplot(4, 4, ax_idx)
evl_p = evp.mix(evp.Flattop(t0=0, tlen=20, w_left=2, w_right=2, amp=1), 0.1377, 0.127, True)
t, wave = evp.decode_envelope(evl_p)
plt.plot(np.arange(t, t+len(wave)*wc_p.resolution, wc_p.resolution), wave.real, '.-', label='py')
evl_c = evc.mix(evc.Flattop(t0=0, tlen=20, w_left=2, w_right=2, amp=1), 0.1377,
                0.127, True)
evl_c = pickle.loads(pickle.dumps(evl_c))
t, wave = evc.decode_envelope(evl_c, wc_c)
wave = np.array(wave, copy=False)
plt.plot(np.arange(t, t+len(wave)*wc_c.resolution(), wc_c.resolution()), wave.real, '.-', label='c++')
plt.legend()
plt.title(evl_p.__class__.__name__+' real')
ax_idx += 1

fig.add_subplot(4, 4, ax_idx)
evl_p = evp.mix(evp.Flattop(t0=0, tlen=20, w_left=2, w_right=2, amp=1), 0.1377, 0.127, True)
t, wave = evp.decode_envelope(evl_p)
plt.plot(np.arange(t, t+len(wave)*wc_p.resolution, wc_p.resolution), wave.imag, '.-', label='py')
evl_c = evc.mix(evc.Flattop(t0=0, tlen=20, w_left=2, w_right=2, amp=1), 0.1377,
                0.127, True)
evl_c = pickle.loads(pickle.dumps(evl_c))
t, wave = evc.decode_envelope(evl_c, wc_c)
wave = np.array(wave, copy=False)
plt.plot(np.arange(t, t+len(wave)*wc_c.resolution(), wc_c.resolution()), wave.imag, '.-', label='c++')
plt.legend()
plt.title(evl_p.__class__.__name__+' imag')
ax_idx += 1

plt.tight_layout()
```
